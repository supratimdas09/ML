```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df.head(3)
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']], df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

```python
import pickle
```

```python
with open('model_pickle', 'wb') as file:
    pickle.dump(model, file)
```

```python
with open('model_pickle', 'rb') as f:
    mp = pickle.load(f)
```

```python
mp.predict([[5000]])
```

```python
#more object which on objects that carry large numpy arrays internally as is often the case (Joblib )
```

```python
from sklearn.externals import joblib
joblib.dump(model, 'model_joblib')
```

```python
mj = joblib.load('model_joblib')
mj.coef_
```
